```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("poznan.csv", encoding= "ansi", sep = ";", error_bad_lines=False)
```

```python
data = data.drop(['miasto.1', 'odzialy'], 1)
```

```python
data_grouped = data.groupby(["Miejscowość", "Ulica", "Nr domu"]).sum()
```

```python
data_grouped.to_csv("data_schools.csv", sep = ";")
```
